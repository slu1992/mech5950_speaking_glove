```python
import os
import cPickle
from gtts import gTTS
from pygame import mixer # Load the required library

f = open('model_bak.pkl', 'rb')
trained = cPickle.load(f)
f.close()

models = trained['models']
clf = models['svm']

def read_text(text):
    tts = gTTS(text, lang='en')
    tts.save("temp.mp3")
    os.system("mpg321 temp.mp3")

    mixer.init()
    mixer.music.load("temp.mp3")
    mixer.music.play()
    os.remove("temp.mp3")

read_text("Program started!")

import numpy as np
import serial
from sklearn import linear_model
from sklearn import neighbors
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier  
from sklearn.ensemble import GradientBoostingClassifier

arduino = serial.Serial('/dev/cu.usbmodem1411', 9600)

count = 5
text = ''

while(1):
    if(arduino.inWaiting()>0):
        data = arduino.readline()
        if(len(data) > 5):
            np_array = np.fromstring(data[1:-1], sep=',').reshape(1, -1)
            #print np_array.shape
            letter = clf.predict(np_array)
            
            letter = str(letter)[2:-2]
            print 'Signals: ' + data + 'Input: ' + letter
            
            if (letter == 'enter'):
                if (len(text)!=0):
                    read_text(text)
                text = ''
                letter = ''
            elif letter == 'delete':
                text = text[:-1]
                letter = ''
            elif letter == 'space':
                text = text + ' '
                letter = ''
            else:
                text += str(letter)
                letter = ''
                
            print 'Text: '+ ''.join(text)
            print
```
