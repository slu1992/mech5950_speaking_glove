### 01. g

```python
import serial

arduino = serial.Serial('/dev/cu.usbmodem1411', 9600)

count = 255
name = 'dataset/g'
datalist = []
while(1):
    if(arduino.inWaiting()>0):
        data = arduino.readline()
        if(len(data) > 5):
            datalist.append(data)
            print data.strip('\n')
            count -= 1
    if count == 0:
        f = open(name+'.txt', 'w')
        f.writelines(["%s" % item  for item in datalist])
        f.close()
        print 'finished!'
        break
```
